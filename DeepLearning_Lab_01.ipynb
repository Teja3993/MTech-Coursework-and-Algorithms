```python
# Perceptron implementation
# with bias term
import random
def perceptron(x,wt):
  f = x[0]
  for i,w in zip(x[1:],wt):
    f = f + (w*i)

  if f>=1:
    return 1
  else:
    return 0

x = [1 ,7.8, 2, 3, 6]
# wt = [1,1,1,1]
wt = random.sample(range(-2,2),4)
print("Random weights are ", wt)

print(perceptron(x,wt))
```

```python
# Perceptron implementation
# with updated weights
import random
def perceptron(x,wt):
  f = x[0]
  for i,w in zip(x[1:],wt):
    w = w /10
    f = f + (w*i)

  if f>=1:
    return 1
  else:
    return 0

x = [1 , 7.8, 2, 3, 6]
# wt = [1,1,1,1]
wt = random.sample(range(-2,2),4)
print("Random weights are ", wt)

print(perceptron(x,wt))
```
