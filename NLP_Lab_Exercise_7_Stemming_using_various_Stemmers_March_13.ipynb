```python
import nltk
from nltk.stem import PorterStemmer, LancasterStemmer, RegexpStemmer, SnowballStemmer

def apply_stemmer(stemmer_name, stemmer_obj, words):
    """Helper function to apply a given stemmer to a list of words and print results."""
    print(f"\n--- {stemmer_name} ---")
    for word in words:
        stem = stemmer_obj.stem(word)
        print(f"{word} ---> {stem}")

def run_default_experiment():
    """Runs the original experiment with the default datasets from the PDF."""
    print("\n=== RUNNING DEFAULT EXPERIMENT (PDF DATA) ===")

    porter = PorterStemmer()
    apply_stemmer("1. Porter Stemmer (Standard)", porter,
                  ['Connects', 'Connecting', 'Connections', 'Connected', 'Connection', 'Connectings', 'Connect'])

    snowball = SnowballStemmer(language='english')
    apply_stemmer("2. Snowball Stemmer (Porter 2.0)", snowball,
                  ['generous', 'generate', 'generously', 'generation'])

    lancaster = LancasterStemmer()
    apply_stemmer("3. Lancaster Stemmer (Aggressive)", lancaster,
                  ['eating', 'eats', 'eaten', 'puts', 'putting'])

    regexp = RegexpStemmer('ing$|s$|able$', min=4)
    apply_stemmer("4. Regexp Stemmer (Custom: -ing, -s, -able)", regexp,
                  ['mass', 'was', 'bee', 'computer', 'advisable'])
    print("=============================================\n")

def get_custom_words():
    """Prompts the user for a custom string and converts it to a list of words."""
    user_input = input("\nEnter words separated by spaces: ").strip()
    if not user_input:
        print("[!] Input cannot be empty.")
        return []
    # Split by spaces to create a list of words
    return user_input.split()

def main_menu():
    """Displays the interactive menu and handles choices."""
    porter = PorterStemmer()
    snowball = SnowballStemmer(language='english')
    lancaster = LancasterStemmer()
    regexp = RegexpStemmer('ing$|s$|able$', min=4) # Custom rules from your PDF

    while True:
        print("\n=== STEMMING ALGORITHMS MENU ===")
        print("1. Run Default Experiment (PDF Data)")
        print("2. Custom Input - Porter Stemmer")
        print("3. Custom Input - Snowball Stemmer")
        print("4. Custom Input - Lancaster Stemmer")
        print("5. Custom Input - Regular Expression Stemmer")
        print("6. Custom Input - Compare ALL Stemmers")
        print("7. Exit")

        choice = input("Select an option (1-7): ").strip()

        if choice == '1':
            run_default_experiment()

        elif choice in ['2', '3', '4', '5', '6']:
            words = get_custom_words()
            if not words:
                continue

            if choice == '2':
                apply_stemmer("Porter Stemmer", porter, words)
            elif choice == '3':
                apply_stemmer("Snowball Stemmer", snowball, words)
            elif choice == '4':
                apply_stemmer("Lancaster Stemmer", lancaster, words)
            elif choice == '5':
                apply_stemmer("Regexp Stemmer", regexp, words)
            elif choice == '6':
                print("\n=== COMPARING ALL STEMMERS ===")
                apply_stemmer("Porter Stemmer", porter, words)
                apply_stemmer("Snowball Stemmer", snowball, words)
                apply_stemmer("Lancaster Stemmer", lancaster, words)
                apply_stemmer("Regexp Stemmer (Rules: -ing, -s, -able | min=4)", regexp, words)
                print("==============================")

        elif choice == '7':
            print("\nExiting program. Goodbye!")
            break

        else:
            print("\n[!] Invalid selection. Please enter a number between 1 and 7.")

# Start the application
if __name__ == "__main__":
    main_menu()
```
