```python
# Import the specific tool from the NLTK library
from nltk import ngrams

def n_gram_model():
    #
    # METHOD 1: Hardcoded Input
    #
    print("--- Method 1 (Fixed Input) ---")

    # 1. Define the string
    gvn_str = "Hello Welcome to Python Programs"

    # 2. Define N (We want Bigrams, so N=2)
    gvn_n_val = 2

    # 3. Split the string into a list of words
    # "Hello Welcome..." -> ['Hello', 'Welcome', 'to', 'Python', 'Programs']
    splt_lst = gvn_str.split()

    # 4. Generate the N-Grams
    # The 'ngrams' function creates a generator object
    rslt_n_grms = ngrams(splt_lst, gvn_n_val)

    # 5. Loop through and print each gram
    for itr in rslt_n_grms:
        print(itr)



    #
    # METHOD 2: User Input
    #
    print("--- Method 2 (User Input) ---")

    # 1. Take input from the user
    # Try typing: "The quick brown fox jumps"
    gvn_str = input("Enter a sentence: ")

    # 2. Take N from the user (e.g., 3 for Trigrams)
    try:
        gvn_n_val = int(input("Enter value for N (e.g., 2 or 3): "))

        # 3. Split and Generate
        splt_lst = gvn_str.split()

        # ERROR CHECK: If text is shorter than N, it won't work
        if len(splt_lst) < gvn_n_val:
            print(f"Error: Your sentence is too short for {gvn_n_val}-grams.")
        else:
            rslt_n_grms = ngrams(splt_lst, gvn_n_val)
            for itr in rslt_n_grms:
                print(itr)

    except ValueError:
        print("Please enter a valid number for N.")

# --- Driver Code ---
if __name__ == "__main__":
    n_gram_model()
```
