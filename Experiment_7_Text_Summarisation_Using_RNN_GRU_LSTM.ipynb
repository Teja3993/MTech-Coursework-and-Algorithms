**SimpleRNN Implementation**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, SimpleRNN, Dense, Embedding

# --- 1. Dimensions & Hyperparameters ---
vocab_size = 100
embedding_dim = 16
latent_dim = 32
seq_length = 5
num_samples = 2000 # Increased to give the model enough examples to learn

# --- 2. Generate Dummy Data (Sequence Reversal) ---
# Generate random sequences of 5 numbers
X_enc = np.random.randint(1, vocab_size, size=(num_samples, seq_length))

# Target: The exact reverse of the input
y_dec = np.flip(X_enc, axis=1)

# Decoder Input: Shifted right by 1, starting with a <START> token (0)
X_dec = np.zeros((num_samples, seq_length))
X_dec[:, 0] = 0
X_dec[:, 1:] = y_dec[:, :-1]

# --- 3. Build SimpleRNN Model ---
# Encoder
encoder_inputs = Input(shape=(None,))
enc_emb = Embedding(vocab_size, embedding_dim)(encoder_inputs)
encoder_rnn = SimpleRNN(latent_dim, return_state=True)
encoder_outputs, state_h = encoder_rnn(enc_emb) # RNN has 1 internal state

# Decoder
decoder_inputs = Input(shape=(None,))
dec_emb = Embedding(vocab_size, embedding_dim)(decoder_inputs)
decoder_rnn = SimpleRNN(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _ = decoder_rnn(dec_emb, initial_state=[state_h])

# Output Layer
decoder_dense = Dense(vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Compile
model_rnn = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model_rnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# --- 4. Train the Model ---
print("--- Training SimpleRNN to Reverse Sequences ---")
# Training for 30 epochs so the loss drops and it learns the logic
model_rnn.fit([X_enc, X_dec], y_dec, batch_size=32, epochs=30, verbose=1)

# --- 5. Test Prediction ---
print("\n--- Output Demonstration ---")
# Grab a random sample from our generated data
sample_idx = 0
sample_enc = X_enc[sample_idx:sample_idx+1]
sample_dec = X_dec[sample_idx:sample_idx+1]

# Ask the model to predict
prediction = model_rnn.predict([sample_enc, sample_dec], verbose=0)
predicted_tokens = np.argmax(prediction[0], axis=-1)

# Display the results
print(f"Input Sequence         : {sample_enc[0]}")
print(f"Expected Target        : {y_dec[0].astype(int)}")
print(f"AI Predicted Sequence  : {predicted_tokens}")
```

**LSTM Implementation**

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Embedding

# --- 1. Dimensions & Hyperparameters ---
vocab_size = 100
embedding_dim = 16
latent_dim = 32
seq_length = 5
num_samples = 2000 # Increased to give the model enough examples to learn the pattern

# --- 2. Generate Dummy Data (Sequence Reversal) ---
# Generate random sequences of 5 numbers
X_enc = np.random.randint(1, vocab_size, size=(num_samples, seq_length))

# Target: The exact reverse of the input
y_dec = np.flip(X_enc, axis=1)

# Decoder Input: Shifted right by 1, starting with a <START> token (0)
X_dec = np.zeros((num_samples, seq_length))
X_dec[:, 0] = 0
X_dec[:, 1:] = y_dec[:, :-1]

# --- 3. Build LSTM Model ---
# Encoder
encoder_inputs = Input(shape=(None,))
enc_emb = Embedding(vocab_size, embedding_dim)(encoder_inputs)
encoder_lstm = LSTM(latent_dim, return_state=True)
encoder_outputs, state_h, state_c = encoder_lstm(enc_emb) # LSTM has 2 internal states

# Decoder
decoder_inputs = Input(shape=(None,))
dec_emb = Embedding(vocab_size, embedding_dim)(decoder_inputs)
decoder_lstm = LSTM(latent_dim, return_sequences=True, return_state=True)
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=[state_h, state_c])

# Output Layer
decoder_dense = Dense(vocab_size, activation='softmax')
decoder_outputs = decoder_dense(decoder_outputs)

# Compile
model_lstm = Model([encoder_inputs, decoder_inputs], decoder_outputs)
model_lstm.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# --- 4. Train the Model ---
print("--- Training LSTM to Reverse Sequences ---")
# Training for 30 epochs so the loss drops and it learns the logic
model_lstm.fit([X_enc, X_dec], y_dec, batch_size=32, epochs=30, verbose=1)

# --- 5. Test Prediction ---
print("\n--- Output Demonstration ---")
# Grab a random sample from our generated data
sample_idx = 0
sample_enc = X_enc[sample_idx:sample_idx+1]
sample_dec = X_dec[sample_idx:sample_idx+1]

# Ask the model to predict
prediction = model_lstm.predict([sample_enc, sample_dec], verbose=0)
predicted_tokens = np.argmax(prediction[0], axis=-1)

# Display the results
print(f"Input Sequence         : {sample_enc[0]}")
print(f"Expected Target        : {y_dec[0].astype(int)}")
print(f"AI Predicted Sequence  : {predicted_tokens}")
```
