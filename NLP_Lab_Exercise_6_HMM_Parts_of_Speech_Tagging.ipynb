```python
# --- Experiment 6: HMM Part-of-Speech Tagging (FIXED) ---

# 1. Import necessary libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

# 2. Download ALL necessary NLTK resources
# We include both the old and new tagger names to ensure compatibility
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng') # <--- This is the fix for your error

def run_pos_tagging():
    print("--- POS Tagging Output ---\n")

    # 3. Define the input text
    txt = ("The Natural Language Toolkit NLTK is a platform used for building programs "
           "for text analysis. One of the more powerful aspects of the NLTK module is the "
           "Part of Speech tagging.")

    # 4. Get English stopwords
    stop_words = set(stopwords.words('english'))

    # 5. Tokenize the text into sentences
    tokenized_sentences = sent_tokenize(txt)

    # 6. Process each sentence
    for sentence in tokenized_sentences:
        # Tokenize sentence into words
        wordsList = word_tokenize(sentence)

        # Remove stopwords
        cleaned_words = [w for w in wordsList if w not in stop_words]

        # 7. Apply POS Tagging
        # The pos_tag function will now automatically find the downloaded resource
        tagged = nltk.pos_tag(cleaned_words)

        print(tagged)

# Run the function
run_pos_tagging()
```

```python
# 1. Import necessary libraries
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

# 2. Download ALL necessary NLTK resources
# Added quiet=True so the terminal doesn't get cluttered with download messages every run
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)
nltk.download('averaged_perceptron_tagger_eng', quiet=True)

def run_pos_tagging(text_input):
    """Processes text and prints POS tags."""
    print("\n--- POS Tagging Output ---")

    # Get English stopwords
    stop_words = set(stopwords.words('english'))

    # Tokenize the text into sentences
    tokenized_sentences = sent_tokenize(text_input)

    # Process each sentence
    for sentence in tokenized_sentences:
        # Tokenize sentence into words
        wordsList = word_tokenize(sentence)

        # Remove stopwords
        cleaned_words = [w for w in wordsList if w not in stop_words]

        # Apply POS Tagging
        tagged = nltk.pos_tag(cleaned_words)

        print(tagged)
    print("--------------------------\n")

def main_menu():
    """Displays the menu and handles user input."""
    default_txt = ("The Natural Language Toolkit NLTK is a platform used for building programs "
                   "for text analysis. One of the more powerful aspects of the NLTK module is the "
                   "Part of Speech tagging.")

    while True:
        print("=== POS Tagging Menu ===")
        print("1. Run with default text (Experiment 6)")
        print("2. Enter custom text")
        print("3. Exit")

        choice = input("Select an option (1-3): ").strip()

        if choice == '1':
            print("\nProcessing default text...")
            run_pos_tagging(default_txt)

        elif choice == '2':
            custom_txt = input("\nEnter the text you want to tag: ").strip()
            if custom_txt:
                run_pos_tagging(custom_txt)
            else:
                print("\n[!] Error: Text cannot be empty. Please try again.\n")

        elif choice == '3':
            print("\nExiting program. Goodbye!")
            break

        else:
            print("\n[!] Invalid selection. Please enter 1, 2, or 3.\n")

# Run the menu-driven application
if __name__ == "__main__":
    main_menu()
```
