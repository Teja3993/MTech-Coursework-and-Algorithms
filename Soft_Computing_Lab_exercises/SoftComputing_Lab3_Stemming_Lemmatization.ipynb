```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# 1. Download the dictionary data (Only need to run these once)
nltk.download('wordnet')
nltk.download('omw-1.4')

def compare_stemming_lemmatization(words):
    # 2. Initialize the tools
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()

    print(f"{'Original':<12} | {'Stemming':<12} | {'Lemmatization'}")
    print("-" * 45)

    for word in words:
        # 3. Apply Stemming
        root_stem = stemmer.stem(word)

        # 4. Apply Lemmatization (treating word as a verb 'v')
        root_lemma = lemmatizer.lemmatize(word, pos='v')

        print(f"{word:<12} | {root_stem:<12} | {root_lemma}")

# Driver Code
word_list = ["running", "flies", "better", "cats", "happily"]
compare_stemming_lemmatization(word_list)
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

# --- SETUP ---
# Download necessary NLTK data files
# (You only need to run these lines once)
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt')  # Often needed for tokenization

def experiment_3_stemming_lemmatization():
    # ==============================
    # PART 1: STEMMING
    # ==============================
    print("--- STEMMING ---")

    # Initialize the Porter Stemmer
    ps = PorterStemmer()

    # List of words from your manual
    words = ["gitam", "university", "sunflower", "beautiful"]

    # Loop through each word and stem it
    for w in words:
        root_word = ps.stem(w)
        print(f"Original: {w:<12} -> Stemmed: {root_word}")

    print("\n" + "="*30 + "\n")

    # ==============================
    # PART 2: LEMMATIZATION
    # ==============================
    print("--- LEMMATIZATION ---")

    # Initialize the WordNet Lemmatizer
    lemmatizer = WordNetLemmatizer()

    # Test word from your manual
    test_word = "foods"

    # Lemmatize the word
    # Note: lemmatize() defaults to noun (n), which works for 'foods' -> 'food'
    lemma_word = lemmatizer.lemmatize(test_word)

    print(f"Original: {test_word:<12} -> Lemmatized: {lemma_word}")

# --- Driver Code ---
if __name__ == "__main__":
    experiment_3_stemming_lemmatization()
```
