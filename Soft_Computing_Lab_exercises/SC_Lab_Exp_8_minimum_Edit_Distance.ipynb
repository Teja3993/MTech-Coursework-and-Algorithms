```python
def min_edit_distance(source, target):
    n = len(source)
    m = len(target)

    # Create a distance matrix D[n+1][m+1] initialized to 0
    D = [[0 for _ in range(m + 1)] for _ in range(n + 1)]

    # Initialization: the zeroth row and column is the distance from the empty string
    # D[0,0] = 0 is naturally handled by the initialization above

    # Cost of Insertion: 1, Deletion: 1
    for i in range(1, n + 1):
        D[i][0] = D[i-1][0] + 1  # del-cost

    for j in range(1, m + 1):
        D[0][j] = D[0][j-1] + 1  # ins-cost

    # Recurrence relation:
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            # sub-cost is 0 if characters are identical, otherwise 1
            if source[i-1] == target[j-1]:
                sub_cost = 0
            else:
                sub_cost = 1

            D[i][j] = min(
                D[i-1][j] + 1,          # Deletion: D[i-1,j] + del-cost
                D[i-1][j-1] + sub_cost, # Substitution: D[i-1,j-1] + sub-cost
                D[i][j-1] + 1           # Insertion: D[i,j-1] + ins-cost
            )

    # Termination
    return D[n][m]

# Main execution matching the expected output
if __name__ == "__main__":
    # Read source and target strings
    source = "adceg"
    target = "abcfg"

    print(f"Source string: {source}")
    print(f"Target string: {target}")

    # Calculate distance
    result = min_edit_distance(source, target)

    # Print output
    print(f"Minimum number of Operations required: {result}")
```

```python
def min_edit_distance_memo(source, target):
    # Dictionary to store the results of overlapping subproblems
    memo = {}

    def dp(i, j):
        # 1. Check if the subproblem has already been solved
        if (i, j) in memo:
            return memo[(i, j)]

        # 2. Base Cases
        if i == 0:
            return j  # If source is empty, cost is inserting all remaining target characters
        if j == 0:
            return i  # If target is empty, cost is deleting all remaining source characters

        # 3. Recurrence Relation
        if source[i-1] == target[j-1]:
            # Characters match: sub-cost is 0, move on to the next characters
            ans = dp(i-1, j-1)
        else:
            # Characters differ: find the minimum cost among the three operations
            ans = min(
                dp(i-1, j) + 1,    # Deletion
                dp(i-1, j-1) + 1,  # Substitution
                dp(i, j-1) + 1     # Insertion
            )

        # 4. Save the computed answer to the memo dictionary before returning
        memo[(i, j)] = ans
        return ans

    # Start the recursive calls from the ends of both strings
    return dp(len(source), len(target))

# Main execution matching your document's test case
if __name__ == "__main__":
    source = "adceg"
    target = "abcfg"

    print(f"Source string: {source}")
    print(f"Target string: {target}")

    # Calculate distance using memoization
    result = min_edit_distance_memo(source, target)

    # Print output
    print(f"Minimum number of Operations required: {result}")
```
