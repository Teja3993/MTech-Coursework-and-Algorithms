```python
import random

# Problem: Maximize number of 1s in a binary string
GENE_LENGTH = 20
POP_SIZE = 20
GENERATIONS = 50
MUTATION_RATE = 0.01
TOURNAMENT_SIZE = 3

# New Constants for Best Practices
CROSSOVER_RATE = 0.8
ELITISM_COUNT = 1

# Helper function to calculate fitness (maximize number of 1s)
def calculate_fitness(chromosome):
    return sum(chromosome)

# Create random individual (Now stores fitness to avoid recalculating)
def create_individual():
    chromosome = [random.randint(0, 1) for _ in range(GENE_LENGTH)]
    return {"chromosome": chromosome, "fitness": calculate_fitness(chromosome)}

# Create initial population
def create_population():
    return [create_individual() for _ in range(POP_SIZE)]

# Selection (Tournament Selection)
def selection(population):
    tournament = random.sample(population, TOURNAMENT_SIZE)
    # Sort based on the pre-calculated fitness stored in the dictionary
    tournament.sort(key=lambda x: x["fitness"], reverse=True)
    return tournament[0]

# Crossover (Single-point crossover)
def crossover(parent1, parent2):
    point = random.randint(1, GENE_LENGTH - 1)

    child1_chrom = parent1["chromosome"][:point] + parent2["chromosome"][point:]
    child2_chrom = parent2["chromosome"][:point] + parent1["chromosome"][point:]

    # Create new dictionary objects for the children
    child1 = {"chromosome": child1_chrom, "fitness": calculate_fitness(child1_chrom)}
    child2 = {"chromosome": child2_chrom, "fitness": calculate_fitness(child2_chrom)}

    return child1, child2

# Mutation (Bit flip)
def mutation(individual):
    chromosome = individual["chromosome"]
    mutated = False

    for i in range(len(chromosome)):
        if random.random() < MUTATION_RATE:
            chromosome[i] = 1 - chromosome[i]
            mutated = True

    # Only recalculate fitness if a mutation actually occurred
    if mutated:
        individual["fitness"] = calculate_fitness(chromosome)

    return individual

# Genetic Algorithm
def genetic_algorithm():
    population = create_population()

    for generation in range(GENERATIONS):
        new_population = []

        # 1. ELITISM: Sort current population and keep the best individual(s)
        population.sort(key=lambda x: x["fitness"], reverse=True)
        # Deep copy the elites so they aren't accidentally mutated later
        elites = [{"chromosome": ind["chromosome"][:], "fitness": ind["fitness"]} for ind in population[:ELITISM_COUNT]]
        new_population.extend(elites)

        # Fill the rest of the new population
        while len(new_population) < POP_SIZE:
            parent1 = selection(population)
            parent2 = selection(population)

            # 2. CROSSOVER RATE: Apply probability to crossover
            if random.random() < CROSSOVER_RATE:
                child1, child2 = crossover(parent1, parent2)
            else:
                # Clone parents if crossover does not occur
                child1 = {"chromosome": parent1["chromosome"][:], "fitness": parent1["fitness"]}
                child2 = {"chromosome": parent2["chromosome"][:], "fitness": parent2["fitness"]}

            # Mutate children
            child1 = mutation(child1)
            child2 = mutation(child2)

            new_population.extend([child1, child2])

        # Ensure population size remains exactly POP_SIZE (in case of odd numbers)
        population = new_population[:POP_SIZE]

        # Get best individual of this generation
        best = max(population, key=lambda x: x["fitness"])
        print(f"Generation {generation+1}: Best Fitness = {best['fitness']}")

        # 3. EARLY STOPPING: Halt if we find the perfect solution
        if best["fitness"] == GENE_LENGTH:
            print(f"\nPerfect solution found at Generation {generation+1}! Stopping early.")
            break

    return max(population, key=lambda x: x["fitness"])

# Run GA
best_solution = genetic_algorithm()
print("\nFinal Best Solution:", best_solution["chromosome"])
print("Final Best Fitness:", best_solution["fitness"])
```
